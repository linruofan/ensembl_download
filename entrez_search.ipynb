```python
from Bio import Entrez
import itertools 
import time
# 提供你的邮箱，这是 NCBI 要求的
Entrez.email = "2540392147@qq.com"

```

```python

def search_literature(keywords):
    # 构建搜索查询
    query = " AND ".join(keywords) 

    try:
        # 执行搜索
        handle = Entrez.esearch(db="pubmed", term=query)
        record = Entrez.read(handle)
        handle.close()

        # 获取文献 ID 列表
        id_list = record["IdList"]

        if id_list:
            print(f"找到 {len(id_list)} 篇包含关键词的文献。")
            print("文献 ID 列表:", id_list)

            # 获取文献摘要
            handle = Entrez.efetch(db="pubmed", id=id_list, rettype="abstract", retmode="text")
            abstracts = handle.read()
            handle.close()
            print("文献摘要:", abstracts)
            return id_list,abstracts
        else:
            print("未找到包含关键词的文献。")

    except Exception as e:
        print(f"发生错误: {e}")


keywords_lists = itertools.product([
'Vibrio',
'Vibrionaceae',
'Vibrionales',
'Gammaproteobacteria',
'Pseudomonadota'],
[
'Mohonavirus',
'Caudoviricetes',
'Uroviricota',
'Heunggongvirae',
'Duplodnaviria'])

# 调用函数进行搜索
for keywords in keywords_lists:
    articles=search_literature(keywords)
    if articles:
        break
    time.sleep(0.5)
articles
    
```
